```python
# \w typically matches [A-Za-z0-9_]

# str.fonksiyonlarını (pandastaki fonksiyonları) uygulayabilmek için öncelikle Series haline getirmek gerekli.
# Aksi takdirde hata verir
```

```python
stop_words = stopwords.words('english')

for i in ["not", "no"]:
        stop_words.remove(i)
        
# olumsuz yorumlari da tesbit edebilmak icin not ve no ifadelerini stopwords listesinden cikardik
```

```python
def cleaning(data):
    
    import re
    # import islemini def in icerisinde yapmak gerekiyor
    
    #1. Removing URLS
    data = re.sub('http\S+', '', data).strip()
    data = re.sub('www\S+', '', data).strip()

    #2. Removing Tags
    data = re.sub('#\S+', '', data).strip()

    #3. Removing Mentions
    data = re.sub('@\S+', '', data).strip()
    
    #4. Removing upper brackets to keep negative auxiliary verbs in text
    data = data.replace("'", "")
    
    # ayraçları kaldırıyorum ki (can't gibi) olumsuzluk içeren kelimeler içinde ayraç içerdiği için stopwords lerde silinmesin.
    # Yorumun olumsuz haliyle de olsa kalması için ilave bir yöntem olarak kullanıldı.
    
    #5. Tokenize
    text_tokens = word_tokenize(data.lower())
    
    #6. Remove Puncs and number
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]
    
    # punctuation ve no lardan kurtuluyoruz.
    
    #7. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]
    
    # stopword lerden temizleme
    # tokenleri teker teker al stopwords ler içinde yoksa olduğu gibi yaz, yani stopwords ise yazma.
    # storwards olmayanları liste içine yazmaya devam et   
    
    #8. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
    
    # stopwardslardan temizlenmiş tokenleri alıyoruz ve lemma uyguluyoryuz
    
    #joining
    return " ".join(text_cleaned)

     # tamamen temizlenmiş ve köklerine inilmiş kelimeleri alıp join yapıyoruz.
```

```python
# def açıklaması:
# - iki tane def te aynı, aşağıdaki bir önceki derste gösterilendir.
# - yukarıdaki def fonksiyonunda ayraçları kaldırma işlemi gerçekleştirdik. upgrade ettik.
# - 1. ayraçlar kaldırıldı ve küçük harfe dönüştürüldü.
# - 2. noktalama işaretler ve numaralardan arındırılma işlemi yapıldı
# - 3. token without punc içindeki tüm tokenleri al ve içinden stop words olmayanlar ile yola devam et.
# - 4. stop word lerden arındırılmış tokenlerin lemma yani köklerine gittik.
# - 5. text imizi birleştirdik.
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# max_df is used for removing terms that appear too frequently, also known as "corpus-specific stop words". For example:

# max_df = 0.50 means "ignore terms that appear in more than 50% of the documents".
# max_df = 25 means "ignore terms that appear in more than 25 documents".
# The default max_df is 1.0, which means "ignore terms that appear in more than 100% of the documents". Thus, the default setting does not ignore any terms.

# min_df is used for removing terms that appear too infrequently. For example:

# min_df = 0.01 means "ignore terms that appear in less than 1% of the documents".
# min_df = 5 means "ignore terms that appear in less than 5 documents".
# The default min_df is 1, which means "ignore terms that appear in less than 1 document". Thus, the default setting does not ignore any terms.
```

```python
vectorizer = CountVectorizer(preprocessor=cleaning, min_df=3) # ngram_range=(1,2), max_features= 1500

# benim corpusum (datasetim) içerisindeki geçen bütün unique tokenlerimi önce tespit eden ve 
# ardından sayan bir işlemdir.

X_train_count = vectorizer.fit_transform(X_train)

# fit ile X_trainde geçen unique tüm tokenleri tespit ediyor ve transform ile onları sayıyor

X_test_count = vectorizer.transform(X_test)

# X_train de eğitildiği için X_train de geçme durumuna göre X_test'te sayma işlemini yapar.
# X_testte geçen bir token, eğitildiği X_train de yok ise onu yok sayar, işlem yapmaz

# min_df = 3    --> corpusumda 3'den daha az document'da (yorumda) geçen tokenleri ignore et.
# max_df = 1000 --> corpusumda 1000'den daha fazla document'da (yorumda) geçen tokenleri ignore et.

# max_features = 1500 --> corpusumda en sık kullanılan ilk 1500 feature (tokeni) dikkate al diğerlerini ignore et.


# Egitime katkisi olmayacak tokenleri almamak icin min_df=3 aldik, 
# yani token sayisi 3 ve daha az olanlari degerlendirmeye almiyoruz.
# max_df=0.5 tüm dökümanlarin %50 sinde geciyorsa ignore et anlamina geliyor. Cok sik tekrar eden token lari elemis oluyoruz.

# ngram_range kelimeleri gruplayip token olarak vektörize islemi uyguluyor. örnegin;
# ngram_range=(1,3) alindiginda ardarda gelen 1'li, 2'li ve 3'lü kelime gruplarini token yaparak sayiyor.
# Birlikte kullanildiginda farkli bir anlam cikan, kelime tamlamalari ya da kaliplarini yakalamak amaci ile ngram_range kullanilir.

# max_features=n ile tokenler icerisinden en fazla kullanilan 1500 tanesini belirleyip vektörize islemi yapiyor.

# demek ki X_train yeterince büyük olmalı ki X_test içinde geçen her türlü olasılığa cevap versin, tüm tokenler için sayma işlemini yapsın, kelime atlamasın. 

# scale işleminde yaptığımız gibi, leakage önlemek için train --> fit_transform, test --> transform
# fit_transform dediğimiz için unique olan tüm kelimeleri sayıyor ve array e dönüştürüyor
# test te geçen bir kelime train de geçmiyorsa model o kelimeyi yok sayar, görmez.
# model ne kadar fazla kelime içeriyorsa o kadar bu handikapın önüne geçiyor.

# FIT train setindeki kelimeleri tespit ediyor, transform sayıyor ve array a dönüştürüyor
# FIT train setine göre yapıldığı için, TEST te TRAİNe göre transform olacak.
```

## Model Comparisons - Vectorization

```python
from sklearn.metrics import confusion_matrix, classification_report, f1_score, recall_score
```

```python
def eval(model, X_train, X_test):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    print(confusion_matrix(y_test, y_pred))
    print("Test_Set")
    print(classification_report(y_test,y_pred))
    print("Train_Set")
    print(classification_report(y_train,y_pred_train))
    
```

## Naive Bayes

```python
# A olayı gerçekleştiğinde B olayının gerçekleşme olasılığı nedir? (P(A|B))
# B olayı gerçekleştiğinde A olayının gerçekleşme olasılığı nedir? (P(B|A))
# Datasetime nasıl uygulayabilirim?
# P(dont | negative) --> dont yardımcı fiili geçtiğinde bunun negatif olma ihtimali nedir
# P(dont enjoy|positive) --> dont ve enjoy tokenleri birlikte geçtiğinde positive olma ihtimali nedir
```

```python
from sklearn.naive_bayes import MultinomialNB, BernoulliNB # BernoulliNB for binary model

# BernoulliNB for binary model
# MultinomialNB, modelimizde birden fazla label var ise bunun kullanılması tavsiye ediliyor.

# dökümanında multinomial ve bernoille den ikisini de deneyip hangisi daha iyi sonuç veriyorsa onunla devam edilmesi tavsiye ediliyor.
# (önceden denenmiş ve MulitnomialNB daha iyi çıkmış.)
```

```python
nb = MultinomialNB(alpha=10)
nb.fit(X_train_count,y_train)

# burada artık sayısallaştırılmış vektörleri kullanıyoruz. text olarak hiçbir şey bırakmıyoruz.
# Alpha arttikca overfittingi engelliyor. Regularization icin alpha katsayisi kullaniliyor.
```

```python
# Aşağıda her model için aldığımız ilkel skorları CROSS VALIDATe yaparak uygun olup olmadıklarını, uyuşup uyuşmadıklarını görebiliriz.
# Bu çalışmada işlemler uzun sürmesin diye modeller için ayrı ayrı CV çalışması yapılmadı
# Sadece seçilen final modelde yapıldı
```

```python
print("NB MODEL")
eval(nb, X_train_count, X_test_count)

# skorlarım dengeli. 
# dolayısıyla overfitting olmadığını anlıyoruz.
# negatifleri iyi yakalamış
# az olmasına rağmen pozitifleri nispeten iyi yakalamış
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score

scoring = {'accuracy': make_scorer(accuracy_score),
            'precision-neg': make_scorer(precision_score, average=None, labels=["negative"]),
            'recall-neg': make_scorer(recall_score, average=None, labels = ["negative"]),
            'f1-neg': make_scorer(f1_score, average=None, labels = ["negative"])}

model = MultinomialNB(alpha=10)
scores = cross_validate(model, X_train_count, y_train, scoring = scoring, cv = 10, return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

```python
from yellowbrick.classifier import PrecisionRecallCurve

viz = PrecisionRecallCurve(
    MultinomialNB(alpha=10),
    classes=nb.classes_,   # label ların isimleri ile gözükmesi için.. aksi halde isim yerine 0,1,2 diye gözükür 
    per_class=True,
    cmap="Set1"
)
viz.fit(X_train_count,y_train)
viz.score(X_test_count, y_test)
viz.show();

# datada dengesizlik (umbalance) söz konusu olduğu için skorlarım gerçekçi olsun diye viz = PrecisionRecallCurve u kullandım.
# data imbalance olduğu için precision recall kullanıldı
```

## TF-IDF

```python
# scale işleminde yaptığımız gibi, leakage önlemek için train --> fit_transform, test --> transform


# fit_transform dediğimiz için unique olan tüm kelimeleri sayıyor, hangi satırlarda geçtiğini buluyor ve array e dönüştürüyor
# test te geçen bir kelime train de geçmiyorsa model o kelimeyi yok sayar, görmez.
# model ne kadar fazla kelime içeriyorsa o kadar bu handikapın önüne geçiyor.

# TF
# FIT train setindeki unique tokenleri tespit ediyor ve hangi satırlarda geçtiğini buluyor, 
# transform bu tokenleri ve geçtiği satırları sayıyor
# FİT train setine göre yapıldığı için, TEST te TRAİNe göre transform olacak.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# eğitim mantığı diğerleriyle tamamen aynı

tf_idf_vectorizer = TfidfVectorizer(preprocessor=cleaning, min_df=3)
X_train_tf_idf = tf_idf_vectorizer.fit_transform(X_train)   # FIT train setindeki unique tokenleri tespit ediyor ve hangi satırlarda geçtiğini buluyor
X_test_tf_idf = tf_idf_vectorizer.transform(X_test)    # TRANSFORM bu tokenlerin sayısını alıyor ve bu tokenlerle TF-IDF yapıyor

# tf_idf: fit deyince X_traindeki tüm unique tokenleri tespit ediyor ve bütün satırlarda (dokümanlarda) o tokenin geçme sıklığını da sayıyor.
# count vectorizer da sadece unique tokenleri tespit ediyordu burda dokümanlardaki tekrar sayısı da işin içinde..

# transform deyince bunları sayıyor ve fonksiyonu uygulayıp tf-idf katsayılarını buluyor.
```

```python
# TF = KELİMEM / TOPLAM KELİME SAYISI
# DF = CORPUS TAKİ TOPLAM DOKÜMAN SAYISI / KELİMEMİN GEÇTİĞİ DOKÜMAN SAYISI
```

```python
# TF(t) = (Number of times term t appears in a document) / (Total number of terms in the document)

# IDF(t) = log(Total number of documents  in a corpus / Number of documents with term t in it)
```

```python
# Example: 
# Consider a document containing 100 words wherein the word cow appears 3 times
# The TF for cow is then (3 / 100) = 0.03 
# Now, let's assume the corpus has 10 million documents and the word cow appears in one thousand of these
# Then, the IDF is calculated as log(10,000,000 / 1,000) = 4
# Thus, the Tf-idf weight is the product of these quantities: 0.03 * 4 = 0.12
```
