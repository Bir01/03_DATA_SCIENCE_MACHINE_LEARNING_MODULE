# Choosing The Optimal Number of Clusters

## Elbow metod

```python
ssd = []

K = range(2,10)   #2 yi sectik, 1 de cok genis alana yayiliyor ve buradan farkli kümeleme cikabilir

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)  #inertia bize bir küme icerisindeki gözlemlerin centroid etrafinda kümelenmesini gösterir,
    #kendi icinde ne kadar siki sikiya kümelendigini gösterir
```

```python
#elbow metod keskin düsüsün durdugu ilk noktayi alir, yani 3 ü alir
```

```python
plt.plot(K, ssd, "bo--")  #b lue, o kümedeki gösteri, --- cizgiler kesik kesik olsun demek
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method") 
```

```python
ssd # sum of squared distance
```

```python
-pd.Series(ssd).diff()   #bu sadece series de calisir,sayisal degerler pozitif ciksin diye basina - koyduk
```

```python
#anlamli düsüsün oldugu noktayi dikkate alir, sonra paralel gitmis 
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show();
```

```python
#yellow brick 4 dedi
```

## Silhouette analysis

```python
from sklearn.metrics import silhouette_score #1 e yakinsa cok iyi kümelenmis, kümeleme sayim düsünce degerim 1 e yaklasmis
#deger negatifse küme sayimi artirmaliyim, 0 a yakinsa küme sayimi azaltmaliyim
```

```python
K_means_model.labels_
```

```python
silhouette_score(X, K_means_model.labels_)  #herbir gözlemin hangi cluester a atandigi bilgisi veriliyor
```

```python
range_n_clusters = range(2,9) #2 den 9 a kadar olan cluester sayilarini veriyorum
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
#silhouette score da 2 en iyi kümeyi verdi, ama 2 cluster yaparsam birinin icinden de kümeler cikar, belli yerde yogunlasma yok
#genis alana yayilip cogunlugu merkeze yakin yerde toplansaydi o zaman 2 yi secebilirdim
# bu nedenle 3 ü seciyorum, 4 e bölsem iyi kümeleme elde etmem, en iyi skor 2 yi haric tutunca 3 cikmis
# bu sayilar bütün gözlemlerin silhoutte skorlarinin ortalamasidir
```

```python
from sklearn.cluster import KMeans

from yellowbrick.cluster import SilhouetteVisualizer

model3 = KMeans(n_clusters=3, random_state=42)
visualizer = SilhouetteVisualizer(model3)

visualizer.fit(X)    # Fit the data to the visualizer
visualizer.poof();
```

```python
model3.labels_
```

```python
visualizer.silhouette_samples_  #datadaki tüm gözlemlerin silhoutte skorlarini döndürür
```

```python
labels = (model3.labels_ == 1)   #1 cluester imi True yapar
```

```python
labels
```

```python
visualizer.silhouette_samples_[labels] # silhoutte scores of 1 classes
# 1 class ina ait sil. skorlarinin ortalamasini döndürdü
```

```python
mean_silhoutte_score_1 = visualizer.silhouette_samples_[labels].mean()
mean_silhoutte_score_1
#1 classindaki tüm sil. skorlarin ortalamasi 0.79 cikti
```

```python
#model3.n_clusters
for i in range(3):
    label = (model3.labels_== i)
    print(f"mean silhouette score for label {i:<4} : {visualizer.silhouette_samples_[label].mean()}")
print(f"mean silhouette score for all labels : {visualizer.silhouette_score_}") #bütün gözlemlerin sil. skorunun ortalamasini döndürür
```
